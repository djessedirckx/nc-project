```python
# dependencies
import gym
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time
import random
%matplotlib inline
```

```python
# testing the environment
env = gym.make('MsPacman-ram-v0')
```

```python
# initiate starting sequence
def intial_sequence(length):

    sequence = []

    for i in range(length):
        n = random.randint(0,8)
        sequence.append(n)
    
    return sequence

# evaluate sequence
def evaluation(action_list, evaluation_per_generation):

    total_reward = 0
    runs = evaluation_per_generation

    for i in range(runs):

            counter = 0
            round_reward = 0
            env.reset()
            done = False
            j = 0

            while j < len(action_list):

                state, reward, done, info = env.step(action_list[j])
                total_reward += reward
                j += 1

                if done:
                    j = len(action_list)
            
            avg_reward = total_reward/runs
        
    print('Reward', avg_reward)

    return avg_reward

# mutate sequence
def mutate(sequence, mutation_prob):

    count = 0 
    for i in range(len(sequence)):

        coin = random.uniform(0, 1)
    
        # if i < 300: (train the first part first?)
        if coin < mutation_prob:

            sequence[i] = random.randint(1,4)
            count += 1
        
    print('percentage changes', count/len(sequence))
    
    return sequence

# main
def run_seq_mutation(length = 1000, eval_per_generation = 5, mutation_prob = 0.7, runs = 10):
    
    # choose mutation probabily random every time (?)

    # collector
    highest_rewards_col = []
    episode_of_change_col = []
    generation_distance_col = []
    reward_col = []
    elapse_time_col = []
    best_generation_action = []

    old_generation = 0

    # load existing sequence (?)
    data = np.load('SQM-1000-prob0.3-action4.npz')
    best_sequence = data['arr_0']

    # initiate sequence + first evaluation
    #best_sequence = intial_sequence(length)
    reward_old = evaluation(best_sequence, eval_per_generation)
    highest_rewards_col.append(reward_old)

    generation = 1
    print('Generation:', generation)

    # main loop
    for i in range(runs):

        t = time.time()

        print('')
        print('Round', i)
        
        # mutation
        # sequence = mutate(best_sequence, mutation_prob)
        sequence = best_sequence
        reward_new = evaluation(sequence, eval_per_generation)

        reward_col.append(reward_new)

        current_generation = i

        elapsed = time.time() - t
        print('Elapse time', round(elapsed, 3))
        elapse_time_col.append(elapsed)

        # check whether new sequence is better than old one
        if reward_new > reward_old:

            best_sequence = sequence   # update weights
            reward_old = reward_new     # update best reward
            generation += 1             # increase generation

            # collect results for plotting
            highest_rewards_col.append(reward_new)
            generation_distance_col.append(current_generation - old_generation)
            episode_of_change_col.append(current_generation)
            best_generation_action.append(sequence)

            # commmand line outputs
            print('')
            print('########## WEIGHT CHANGE ############')
            print('Generation:', generation)
            print('Rounds betw. generations:', current_generation - old_generation)
            
            old_generation = current_generation
    
    #np.savez('SQM-1000-prob0.05-action4.npz', best_sequence)

    return highest_rewards_col, generation_distance_col, reward_col, episode_of_change_col, best_sequence, elapse_time_col, best_generation_action
```

```python
# sequence length = 1000, eval_per_generation = 5, mutation_prob = 0.05, runs = 10

highest_rewards, gen_dis, all_rewards, change_episode, sequence, elapse_time, all_actions = run_seq_mutation(1000, 5, 0.05, 50)
```

```python
mean_reward = sum(all_rewards)/len(all_rewards)

counter = 0
for i in all_rewards:
    counter += (i - mean_reward) ** 2

variance = counter / len(all_rewards)
std = variance ** 0.5

print(all_rewards)

print('mean', mean_reward)
print('variance', variance)
print('std', std)
```

```python
# plotting
# unique, counts = np.unique(all_actions[0], return_counts=True)
# distribution = dict(zip(unique, counts))

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 15})

# Create a figure instance
data_to_plot = []
for i in range(len(all_actions)):
    data_to_plot.append(all_actions[i])

# Create a figure instance
fig = plt.figure(figsize=(8, 12))

# Create an axes instance
ax = fig.add_axes([0,0,1,1])

# Create the boxplot
bp = ax.violinplot(data_to_plot, showmeans = True, showextrema = True, showmedians = True)

for partname in ('cbars','cmins','cmaxes','cmeans','cmedians'):
    vp = bp[partname]
    vp.set_edgecolor('slategray')
    vp.set_linewidth(1)

for pc in bp['bodies']:
    pc.set_facecolor('slategray')
    pc.set_edgecolor('slategray')
    pc.set_alpha(1)

plt.grid()
plt.xlabel('Best Generation') 
plt.ylabel('Probability per action')
start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(0, len(data_to_plot) +1, 1))
plt.title("Action distribution") 
plt.tight_layout()
plt.savefig('SQM/pacman-action-distr-1000-prob0.05-action4.pdf', bbox_inches='tight')
plt.show()

# plot the obtained results
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})

# Reward improvement per generation
plt.figure(figsize=(12, 8))
plt.plot(highest_rewards, color='slategray', marker = 'o', markersize=12)
#plt.style.use('ggplot')
plt.grid()
plt.xticks(range(0, len(gen_dis)+1, 1))
plt.xlabel('Generation') 
plt.ylabel('Obtained reward') 
plt.title("Best reward obtained by each Generation")
plt.savefig('SQM/pacman-reward-1000-prob0.05-action4.pdf')
plt.show() 

# Distance between generations
plt.figure(figsize=(12, 8))
plt.bar(range(1, len(gen_dis)+1, 1), gen_dis, color='slategray')
#plt.style.use('ggplot')
plt.grid()
plt.xticks(range(1, len(gen_dis)+1, 1))
plt.xlabel('Generation')
plt.ylabel('Mutations between Generations')
plt.title("Distance between Generations")
plt.savefig('SQM/pacman-distance-1000-prob0.05-action4.pdf')
plt.show() 


# Plot all rewards

# change red and gray colour?
window=20
xcoords = change_episode

plt.figure(figsize=(12,8))
rolling_mean = pd.Series(all_rewards).rolling(window).mean()
std = pd.Series(all_rewards).rolling(window).std()
plt.grid()
plt.plot(rolling_mean)
plt.fill_between(range(len(all_rewards)),rolling_mean-std, rolling_mean+std, color='orange', alpha=0.4)
# plt.axvline(x=30, color='slategray', linestyle='-')

for xc in zip(xcoords):
    plt.axvline(x=xc, label='line at x = {}'.format(xc), c='slategray')

plt.title('Reward Moving Average ({}-episode window)'.format(window))
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.grid()
plt.savefig('SQM/pacman-reward-over-1000-prob0.05-action4.pdf')
plt.show()


# elapse time
total = sum(elapse_time)

plt.figure(figsize=(12, 8))
plt.plot(elapse_time, color='slategray')
plt.xlabel('Episode') 
plt.ylabel('Time')
plt.title("Elapse time (total " + str(round(total, 2)) + " sec.)" ) 
plt.grid()
plt.savefig('SQM/pacman-elapse-time-1000-prob0.05-action4.pdf', bbox_inches='tight')
plt.show()
```
